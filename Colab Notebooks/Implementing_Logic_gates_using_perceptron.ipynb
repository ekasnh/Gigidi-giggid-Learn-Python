Inside the neurons, the two main important steps happen-

1. Weighted Sum.
2. Activation Function.

The first step is the weighted sum, which means all of the weights assigned to the synapses are added with input values. Something like that-

[ x1.w1+x2.w2+x3.w3+..Xn.Wn]

After calculating the weighted sum, the activation function is applied to this weighted sum. And then the neuron decides whether to send this signal to the next layer or not.

# AND Gate 

```python
import numpy as np
```

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]]) #input values
y = np.array([[0],[0],[0],[1]]) #y values
```

```python
print(x)
```

```python
print(y)
```

```python
#initialize the weights
np.random.seed(20)   ## fix thr random value
weights = -np.random.random((2,1))
weights
```

```python
#calculation of Z, the weighted sum with a bias value of 0.02
sum = np.dot(x,weights) + 0.00001
sum
```

```python
#creating a function for sigmoid activation function
def activation(x):   
  return 1/(1+np.exp(-x))
```

```python
#calculating the predicted value yhat by applying sigmoid activation function on weighted sum z
ypred = activation(sum)
ypred
```

```python
#rounding off the predicted value 
ypred.round()      
```

```python
#calculating the prediction error err=y-yhat
error = y - ypred
error
```

```python
#applying gradient descent to minimize the prediction error
def gradient(a,x):
  return a * x * (1-x)
```

```python

chng_in_weight = gradient(error,ypred)      ## Apply Gradient Descent
weights_new = weights + np.dot(x.T,chng_in_weight)  ## Weight updation
weights_new
```

```python
#### Training 
for i in range(5000):
  sum = np.dot(x,weights) + 0.04
  ypred_n = activation(sum)
  error = y - ypred_n
  chng_in_weight = gradient(error,ypred_n)      ## Apply Gradient Descent
  weights_new = weights + np.dot(x.T,chng_in_weight)
  #if i%50 == 0:
   # print(error,end="\n\n")
weights_new
```

```python
ypred_n
```

```python
ypred.round()
```

```python
weights
```

```python
prediction_error=(y-ypred)
prediction_error
```

```python
np.ceil(prediction_error)
```
